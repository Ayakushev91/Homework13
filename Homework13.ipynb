### Задание 1

```python
"""
Для датафрейма log из материалов занятия создайте столбец source_type
"""
```

```python
import pandas as pd
import re
```

```python
log = pd.read_csv('/Python/visit_log.csv', sep=';')
log.head()
```

```python
# traffic_source без изменений
log['source_type'] = log['traffic_source']
```

```python
#если источник traffic_source равен yandex или google, то в source_type ставится organic
log.loc[log['traffic_source'].str.contains('yandex|google'), 'source_type'] = 'organic'
```

```python
#для источников paid и email из России - ставим ad
log.loc[(log['region'].str.contains('Russia') & log['traffic_source'].str.contains('paid|email')), 'source_type'] = 'ad'
```

```python
#для источников paid и email не из России - ставим other
log.loc[(~log['region'].str.contains('Russia') & log['traffic_source'].str.contains('paid|email')), 'source_type'] = 'other'
```

```python
# выводим резульат
log.head(20)
```

### Задание 3

```python
"""
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp
для данного значения userId.
"""
```

```python
#считываем файл с оценками 
ratings_df = pd.read_csv('/Python/ml-latest-small/ratings.csv')
```

```python
ratings_df.head()
```

```python
#считаем количество оценок и время жизни пользователя
result = (
    ratings_df
        .groupby('userId')
        .agg({'rating': 'count', 'timestamp': lambda x: (x.max()-x.min())})
        .rename(columns = {'timestamp': 'LTV'})
)
```

```python
#отфильтровываем результат по количеству оценок
result.loc[result.rating > 100].head(20)
```
